# Importing Libraraies

```python
import pandas as pd
from urllib.request import Request,urlopen
from bs4 import BeautifulSoup
import requests 
import re
```

```python
hdr = {'User-Agent': 'Mozilla/5.0'}
#lists for Product name, image link and product url
prod_name=[]
img_link=[]
web_url=[]
```

```python
url = input('Enter the URL you want to Scrape: ')#https://yoshops.com
```

# Scraping Multiple Pages

```python
print("IT MAY TAKE SOME TIME TO SCRAPE ALL THE WEBPAGES...... PLEASE WAIT!!!")
for pageno in range(1,12):
    pg = url+'/products?page={}'.format(pageno)
    req = Request(pg , headers = hdr)
    page1 = urlopen(req)
    soup1 = BeautifulSoup(page1)

    prod = soup1.select('.product-title') 
    image = soup1.find_all('div' , class_='product-thumb-inner')
    prod_link = soup1.find_all('a' , class_='product-link')
    
    #For getting all Product Names
    for i in range(len(prod)):
        prod_name.append(prod[i].text)
        
    #For getting all Image Links
    for i in image:
        img_link.append(i.find("img")["src"])
        
    #For getting all Product Links
    for i in prod_link:
        web_url.append("https://yoshops.com/"+i["href"].replace('\n',''))

```

# Converting Scraped Data into .csv File

```python
df = {'Product_Name':prod_name,'Image_Link':img_link,'Web_url':web_url}
dataset = pd.DataFrame(data = df)
```

```python
dataset.to_csv('yoshops.csv') #converting to csv
```
